```python
# Import modules and packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
import librosa as lb
import seaborn as sns
```

## Data extraction

```python
data_dir = './Heartbeats/set_a'
audio_files = glob(data_dir + '/*.wav')
```

```python
y, sr = lb.load(audio_files[10])
time = np.arange(len(y))/sr
```

### Plotting heartbeats

```python
sns.set_palette('coolwarm')
for file in audio_files:
    y, sr = lb.load(file)
    time = np.arange(len(y))/sr
    sns.lineplot(y=y,x=time, linewidth=0.3)
    plt.title('Heartbeat audio plot')
    plt.xlabel('Time [s]')
    plt.ylabel('Amplitude')
    plt.show()
```
